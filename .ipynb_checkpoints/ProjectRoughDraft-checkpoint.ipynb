Flight data from the JFK airport from November 2019 to December 2020.\

```python
df = pd.read_csv('M1_final.csv')
df.head() # quick preview
```

```python
y = df.TAXI_OUT # setting the output column as y
X = df.drop('TAXI_OUT',axis = 1, inplace = True) # dropping the target column
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
df.Wind.fillna('WSW', inplace = True)
df.rename(columns = {'Wind':'Wind_fillna'}, inplace = True)
print("Nan values in 'Wind_fillna':", df.Wind_fillna.isnull().sum())
```

```python
sns.boxplot(data = df, x = 'OP_UNIQUE_CARRIER', y ='TAXI_OUT')
```

Initial observations is that **AS (Alaskan Airlines)** has a higher distribution of taxi_out times.\
The average taxi_out for AS is higher than the rest by a significant margin, compared to the other airlines, which are relatively the same.\
However, there are many outliers for **AA (American Airlines)** and **B6 (Jetblue Airlines)**.\
From these observations, AS, AA, and B6 have significant averages and can be considered as the main candidates for dummy variables within OP_UNIQUE_CARRIER, and we can group the rest of the airlines in an 'OTHER' category.

```python
df.OP_UNIQUE_CARRIER = df.OP_UNIQUE_CARRIER.replace(['DL','MQ','9E','YX','HA','OO'], 'OTHER')
df.groupby('OP_UNIQUE_CARRIER').TAXI_OUT.count()
```
