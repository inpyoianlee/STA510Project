**Motivation and Summary**

This Kaggle dataset is about flights leaving JFK Airport between November 2019 and December 2020. The target variable, 'TAXI_OUT', represents the time on the aircraft runway, which could directly affect the cost of the flight. This data is chosen since the idea of traveling and going on vacations is fun, but the preparation to accomplish all of that is just as important. It is arguable that good preparation can make or break the start of a trip any person looks forward to.

The process of going from station to station to undergo security checks, keeping an eye out for the flight schedule, and waiting for the plane to take-off towards a desired destination are unavoidable. Customers face this pre-travel process every time they want to go anywhere, but what they do not always consider is the behind-the-scenes operations that help move this process along smoothly, or not. 

Many customers have probably thought, 'Why isn't my plane here yet?', or 'We've been waiting on the plane for thirty minutes already. When are we going to leave?' These concerns occur since there are multiple factors happening all at once. The big question is, 'Which factors are strongly correlated to the runway time?' 

This report will answer this question by fitting the best regression model with regards to the F-statistic, the coefficient of determination R2 and its adjusted version, the AIC and BIC, and the mean squared error MSE.

**Separate target and feature variables**

```python
y = df['TAXI_OUT']
X = df.drop('TAXI_OUT', axis = 1)
```

**Converting 'Dew Point' object column to series of integers.**

```python
#Dew Point is actually a string of numbers instead of integers
print(df['Dew Point'])
#use astype(int) to convert to numerical data
df['Dew Point'] = df['Dew Point'].astype(int)
```

**Separate training and test data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

#training data for predictors
X_train
```

```python
#test data for predictors
X_test
```

**Separate categorical and numerical features**

```python
categorical_features = list(X_train.select_dtypes('object').columns)
numerical_features = list(X_train.select_dtypes(exclude = ['object']).columns)

#print list of each feature data type
print('Categorical features:', categorical_features, '\n')
print('Numerical features:', numerical_features)
```

**Pipelines and ColumnTransformer for Simpler Data Preprocessing**

```python
#Use Pipeline and ColumnTransformer to bundle preprocessing steps and create more concise code in one line of code
#These eliminate the need to preprocess (encode/impute) and fit the model separately in several lines

#Only need to scale numerical features (i.e. one method), so no need to bundle preprocessing steps in pipeline
#Change type of scaler?
numerical_transformer = StandardScaler()

#Pipeline (and not ColumnTransformer) because bundling preprocessing steps for the SAME type of columns (i.e. categorical)
#List type of preprocessing in order of desired transformation (i.e. impute missing 'Wind' values first, then one-hot encode)
#Format is list of tuples, two elements each --> ('variable name', preprocessing method)
categorical_transformer = Pipeline(steps = [
    ('imputer', SimpleImputer(strategy = 'most_frequent')), 
    ('onehot_encoder', OneHotEncoder(handle_unknown = 'ignore'))
])

#ColumnTransformer since bundling preprocessing steps for two DIFFERENT column types (i.e. numerical and categorical)
#Format is list of tuples, three elements each --> ('variable name', preprocessing method, type of features to preprocess)
preprocessor = ColumnTransformer(transformers = [
    ('numerical', numerical_transformer, numerical_features), 
    ('categorical', categorical_transformer, categorical_features)
])
```

**Default linear regression model with no hyperparameter tuning:**

```python
#Define default linear model without any hyperparameter tuning  
linreg = LinearRegression()

#Pipeline to combine preprocessing and model fitting for the entire/SAME data
linreg_pipeline = Pipeline(steps = [('preprocessor', preprocessor), ('default linear model', linreg)])
#Fit model using training data and predict with test data
linreg_pipeline.fit(X_train, y_train)
linreg_pred = linreg_pipeline.predict(X_test)

#Print R2 score prediction
print('R2 linreg score:', r2_score(y_test, linreg_pred))
print('R2 score is very low, so a linear model is a poor fit for this data.')
```
